```python
import pandas as pd
import numpy as np
import re
from datetime import datetime
import matplotlib.pyplot as plt
def m2d(x):
    return {
        1: 0,
        2: 31,
        3: 31+29,
        4: 31+29+31,
        5: 31+29+31+30,
        6: 31+29+31+30+31,
        7: 31+29+31+30+31+30,
        8: 31+29+31+30+31+30+31,
        9: 31+29+31+30+31+30+31+31,
        10: 31+29+31+30+31+30+31+31+30,
        11: 31+29+31+30+31+30+31+31+30+31,
        12: 31+29+31+30+31+30+31+31+30+31+30
    }[x]
sizes = [2478  ,259,   97,  208,  178,  100,   93,   79,  158,  351]
useful=[0,2,3,6,7,8]
data = pd.read_csv('kmeans_clusters.csv')
```

```python
groups=[data.loc[data['label_kmeans'] == i] for i in range(0,10)]
times=[group["created_at"] for group in groups]
for idx,group in enumerate(times):
    times[idx]=[' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z ])|(\w+:\/\/\S+)", " ", time).split()) for time in group]
times_as_dt= [np.zeros_like(a) for a in times]
hours_group=[np.zeros_like(a) for a in times]
day_group=[np.zeros_like(a) for a in times]
for idx,group in enumerate(times):   
    times_as_dt[idx]=[datetime.strptime(time, '%Y %m %d %H %M %S') for time in group]
for idx,group in enumerate(times_as_dt):
        hours_group[idx]=[(i.hour*3600+i.minute/60+i.second/3600)/3600.0 for i in group]
        day_group[idx]=[m2d(i.month)+i.day for i in group]

    
    
    
```

```python
hours_toplot=[]
day_toplot=[]
for idx,item in enumerate(hours_group):
    if(idx in useful):
        hours_toplot.append(item)
for idx,item in enumerate(day_group):
    
    if(idx in useful):
        
        day_toplot.append(item)
```
